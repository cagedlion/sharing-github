### 2.Which all variables have an outlier?

```python
sns.set(font_scale=0.7)
plt.figure(figsize=(30,20))
sns.boxplot(data=mydata);
```

```python
sns.set(font_scale=1)
data_plot=mydata[mydata.dtypes[mydata.dtypes!='object'].index]
fig=plt.figure(figsize=(16,8))
for i in range(0,len(data_plot.columns)):
   ax=fig.add_subplot(7,3,i+1)
   sns.boxplot(data_plot[data_plot.columns[i]])
   ax.set_title(data_plot.columns[i],color='Blue')
```

#### Ans2. As can be seen from the above boxplots, all the variables except 'Saturated Fat', 'Saturated Fat (% Daily Value)' and  'Dietary Fiber',  have outliers.

### 3. Which variables have the highest correlation? Plot them and find out the value?

### 6. Which 4 food items contain the most amount of Saturated Fat?

```python
df3=mydata[['Item','Saturated Fat']].groupby('Item').mean().round()
df3
```

```python
df3.sort_values(by=('Saturated Fat'),ascending=False).head(4)
```

#### Ans6. As can be seen from above, the top four items containing saturated fat are 'McFlurry with M&M's Candies (Medium)', 'Big Breakfast with Hotcakes (Large Biscuit), 'Chicken McNuggets (40 piece)' and 'Frappe Chocolate Chip (Large)'.
